### Series

```python
#We would actually do this with pandas, but this is how we would have done it before pandas....
#We won't do things like this with loops when working with pandas..it's much easier
with open('US_Unemployment_Oct2012.csv', 'r', encoding = 'UTF-8') as unemp: #opens file for reading
    data = {} #empty dictionary
    unemp.readline()
    for line in unemp: #looks at file line by line
        #print(line)
        info = line.rstrip().split(',') #removing spaces or /n, split into a list
        #print(info)
        data[info[0]] = float(info[1])
print(data)
```

```python
data_series = pd.Series(data)
data_series
```

```python
print(data_series.index)
print(data_series.values)
```

```python
print(data_series.loc['CA'])
print(data_series.iloc[4])
```

```python
data_series.loc['CA':'ID']
```

```python
data_series.iloc[4:12]
```

```python
data = [10.1, 7.7, 8.4, 7.1]
#How do I divide them all by 100?
for item in data:
    proportions  = item/100
    print(proportions)
```

```python
proportions = data_series/100
proportions.iloc[0:5]
```

```python
data = [10.1, 7.7, 8.4, 7.1]
#How do I cast them all to ints?
rounded = [int(item) for item in data]
print(rounded)
```

```python
rounded = data_series.astype('int')
rounded.iloc[0:5]
```
