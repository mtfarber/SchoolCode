```python
%matplotlib inline
import folium
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cal = pd.read_csv('CalEnviroScreen_2.0_2014.csv') #https://oehha.ca.gov/calenviroscreen/maps-data/download-data
pd.set_option('display.max_columns',100)
cal.head()
```

```python
cal['California County'].value_counts()
```

```python
county = cal.groupby('California County').agg('mean')[['Pollution Burden Score', 'Pollution Burden Pctl']]
county.head()
```

```python
county.index.value_counts();
```

```python
m = folium.Map(location=[39, -98], zoom_start=4)# Add the color for the chloropleth:
folium.Choropleth(geo_data='counties_2010_us_050_00_20m.json',data=county,\
             columns = [county.index,'Pollution Burden Score'],\
             key_on='feature.properties.NAME', \
             fill_color = 'PuBuGn',\
         fill_opacity=0.7,line_opacity=0.2,nan_fill_color='beige' ).add_to(m)
m.save('CA_counties_Poll_mean.html')
```

```python
county.index = county.index.str.rstrip()
```

```python
with open('counties_2010_us_050_00_20m.json', 'r') as counties:
    for i in range(5):
        print(next(counties))
```

```python
with open('counties_2010_us_050_00_20m.json', 'r') as counties:
    with open('Just_CA.json','w') as ca:
        ca.writelines(next(counties))
        ca.writelines(next(counties))
        result_str = ''
        for line in counties:
            if '"STATE": "06"' in line:
                result_str += line + ',\n'    
        result_str = result_str[:-2]
        ca.writelines(result_str)        
        ca.writelines(']\n}\n')
```

```python
m = folium.Map(location=[39, -98], zoom_start=4)# Add the color for the chloropleth:
folium.Choropleth(geo_data='Just_CA.json',data=county,\
             columns = [county.index,'Pollution Burden Score'],\
             key_on='feature.properties.NAME', \
             fill_color = 'PuBuGn',\
         fill_opacity=0.7,line_opacity=0.2, legend_name = 'Burden Score').add_to(m)
m.save('CA_counties_Poll_mean.html')
```
