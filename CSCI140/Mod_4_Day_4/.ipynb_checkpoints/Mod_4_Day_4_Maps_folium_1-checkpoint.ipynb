```python
%matplotlib inline
import folium
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lakes = pd.read_csv('Lakes_Edit_Data.txt', sep = '\t')
lakes.head()
```

```python
FI_COORDINATES = (64.64, 27.02) #How do we get these coordinates
fi_map = folium.Map(location=FI_COORDINATES, zoom_start=5) #What does zoom_start do?
fi_map #You will see nothing if not connected to web and in some browsers (Internet Explorer)
```

```python
Helsinki = [60.167408, 24.9425683]
fi_map_H = folium.Map(location=FI_COORDINATES, zoom_start=5)
H_marker = folium.Marker(location=Helsinki, icon = folium.Icon(color='green',icon='star'),popup='Helsinki')
H_marker.add_to(fi_map_H)
fi_map_H #What does this do?
```

```python
fi_map_2 = folium.Map(location=FI_COORDINATES, zoom_start=5)
coords = zip(lakes['Latitude'],lakes['Longitude'])
for lat,lon in coords:
    cm = folium.CircleMarker(location = [lat,lon], fill_color='blue', radius=3, color='blue')
    cm.add_to(fi_map_2)
folium.Marker(location=Helsinki, icon = folium.Icon(color='green',icon='star'),popup='Helsinki').add_to(fi_map_2)
fi_map_2
```

```python
fi_map_3 = folium.Map(location=FI_COORDINATES, zoom_start=5)
colors = {'Eutrophic':'red','Mesotrophic':'green','Oligotrophic':'blue'}
full_zip = zip(lakes['Latitude'], lakes['Longitude'], list(lakes['Status'].values))
for lat,lon,status in full_zip:
    folium.CircleMarker(location = [lat, lon], fill_color=colors[status], radius=5, color=colors[status]).add_to(fi_map_3)
folium.Marker(location=Helsinki, icon = folium.Icon(color='black',icon='star'),popup='Helsinki').add_to(fi_map_3)

fi_map_3
```
